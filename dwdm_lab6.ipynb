```python
map2 = {'no':0,'yes':1}
df['housing'] = df['housing'].map(map2)
print(df['housing'].head(10))
```

```python
map3 = {'no':0,'yes':1}
df['loan'] = df['loan'].map(map3)
print(df['loan'].head(10))
```

```python
import numpy as np
map4 = {'blue-collar':3,
'management':0,
'technician':1,
'admin.':5,
'services':6,
'retired':4,
'self-employed':7,
'entrepreneur':2,
'unemployed':8,
'housemaid':9,
'student':10,
'unknown':np.nan}
df['job'] = df['job'].map(map4)
print(df['job'].head(10))
```

```python
map5 = {'secondary':1,
'tertiary':0,
'primary':2,
'unknown':np.nan}
df['education'] = df['education'].map(map5)
print(df['education'].head(10))
```

```python
map6 = {'no':0,'yes':1}
df['default'] = df['default'].map(map6)
print(df['default'].head(10))
```

```python
df['contact'] =df['contact'].map({'unknown':np.nan,'telephone':0,'cellular':1})
print(df['contact'].head(10))
```

```python
df['month'] = df['month'].map({
'may': 5,
'jul' :7,
'aug': 8,
'jun' : 6,
'nov' : 11,
'apr' :4,
'feb' : 2,
'jan' : 1,
'oct' :10,
'sep' :9,
'mar' : 3,
'dec' :12
})
print(df['month'].head(10))
```

```python
df['poutcome'] =df['poutcome'].map({'unknown':np.nan,'failure':0,'other':1,'success':2})
print(df['poutcome'].head(10))
```

```python
df['y'] = df['y'].map({'no':0,'yes':1})
print(df['y'].head(10))
```

```python
df['duration'].plot(kind = 'hist')
```

```python
df_min_max_scaled = df.copy()
a = 0
b = 1
column = 'duration'
df_min_max_scaled[column] = a + (df_min_max_scaled[column] -
df_min_max_scaled[column].min()) * (b - a)/(df_min_max_scaled[column].max() - df_min_max_scaled[column].min())
print(df_min_max_scaled)
```

```python
df_min_max_scaled['duration'].plot(kind = 'hist')
```

```python
df_z_scaled = df.copy()
column = 'duration'
df_z_scaled[column] = (df_z_scaled[column] -
df_z_scaled[column].mean()) / df_z_scaled[column].std()
print(df_z_scaled)
```

```python
df_z_scaled['duration'].plot(kind = 'hist')
```
