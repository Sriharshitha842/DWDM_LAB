```python
import numpy as np
import time
start = time.time()
matrix = np.random.randint(9, size=(100,100))
print(matrix)
matrix1 = np.random.randint(9, size=(100,100))
print(matrix1)
prod = np.dot(matrix, matrix1)
print(prod)
end = time.time()
Execution_time_1 = end-start
```

```python
import numpy as np
start = time.time()
result = np.zeros([100,100])
for i in range(len(matrix)):
  for j in range(len(matrix1[0])):
    for k in range(len(matrix1)):
      result[i][j]+=matrix[i][k]*matrix1[k][j]
for r in result:
  print(r)
end = time.time()
Execution_time_2 = end-start
```

```python
#Execution time using numpy
print(Execution_time_1)
#execution time using for loop
print(Execution_time_2)
```
