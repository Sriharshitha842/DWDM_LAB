```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/data.csv')
df.head()
```

```python
x = df['Rank'][:10]
y = df['Global_Sales'][:10]
plt.plot(x,y)
plt.xlabel('Game Rank')
plt.ylabel('Game Global Sales')
plt.title('Game Gloabal Sales - Top Ten')
plt.legend(['Rank vs Sales'])
plt.show()
```

```python
plt.plot(x,y,'r',linestyle='steps',linewidth=5)
plt.legend(['Rank vs Sales'])
plt.xlabel('Game Rank')
plt.ylabel('Game Global Sales')
plt.title('Game Gloabal Sales - Top Ten')
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)
plt.scatter(x,y,color='red')
plt.legend(['Rank vs Sales'])
plt.xlabel('Game Rank')
plt.ylabel('Game Global Sales')
plt.title('Game Gloabal Sales - Top Ten')
plt.show()
```

```python
#2.1
x = df['Year']
y = df['Global_Sales']
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)
plt.bar(x,y,color='maroon')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Global Sales Per Year')
plt.show()
```

```python
#2.2
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(6)
df.pivot_table(values = 'Global_Sales', index = 'Platform', aggfunc='sum').plot(kind='bar',color='maroon')
plt.title('Global Sales by platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales ')
plt.tight_layout()
plt.show()
```

```python
#2.3
x = df['Platform']
y = df['Global_Sales']
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(6)
plt.bar(x,y,color='maroon')
plt.xlabel('Platform')
plt.ylabel('Global Sales')
plt.title('Global Sales by Platform')
plt.show()
```
